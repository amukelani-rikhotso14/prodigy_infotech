```python
import pandas as pd

file=r"C:\Users\Amukelani\Downloads\titanic\titanic.csv"
df=pd.read_csv(file)
# print(df.to_string())
# print(df.head())
# print(df.info())
# df.describe()
df.head()
```

```python
# cleaning the data by handling missing values

# print(df.isnull().sum())

#fill missing age values using the median
df['age'].fillna(df['age'].median(), inplace=True)

df['fare'].fillna(df['fare'].median(), inplace=True)


print(df.isnull().sum())
```

```python
#Exlaratory data analysis
#visualize the distribution of numerial features
import matplotlib.pyplot as plt
import seaborn as sns

#visualise the age frequency
plt.figure(figsize=(20,12))
df_age=df['age']
sns.displot(df_age,kde=True)
plt.title("The Age distribution on the titanic")
plt.xlabel("Age")
plt.ylabel("frequency")

plt.show()
```

```python
#visualize the categorical features

#visualize the sex
df_sex=df['sex'].value_counts().reset_index()
print(df_sex)
sns.barplot(x='sex',y='count', data=df_sex)
plt.title("Sex distribution of passengers in the titanic")
plt.xlabel("sex")
plt.ylabel("count")
plt.show()

#visualize the emabrked
df_embarked=df['embarked'].value_counts().reset_index()
sns.barplot(x='embarked', y='count',data=df_embarked)
plt.xlabel("embarked")
plt.ylabel("count")
plt.title("embarked count")
plt.show()

#frequency of class
df_class=df['pclass'].value_counts().reset_index()
sns.barplot(x='pclass',y='count',data=df_class)
plt.xlabel("passenger class")
plt.ylabel("count")
plt.title("passenger class count")
plt.show()

#surved
df_survived=df['survived'].value_counts().reset_index()
sns.barplot(x='survived',y='count',data=df_survived)
plt.xlabel("survived")
plt.ylabel("count")
plt.title("count of survived passengers")

#parch
df_parch=df['parch'].value_counts().reset_index()
sns.barplot(x='parch',y='count',data=df_parch)
plt.xlabel("parch")
plt.ylabel("count")
plt.title("the parch count")

#sibsip
df_sib=df['sibsp'].value_counts().reset_index()
sns.barplot(x='sibsp',y='count',data=df_sib)
plt.xlabel("sibsp")
plt.ylabel("count")
plt.title("the sibsp count")
```

```python
#relationship between embarked point and survival rate
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='survived',hue='embarked',data=df)
plt.xlabel("embarked point")
plt.ylabel("sex of passengers")
plt.title("relationship between embarked point and sex")
plt.show()

#relationship between the survived and sex
sns.countplot(x='survived',hue='sex',data=df)
plt.xlabel("survival rate of passengers")
plt.ylabel("sex of passengers")
plt.title("relationship survived and sex")
plt.show()

# #relationship between pclass and survived
# sns.countplot(x='survived',hue='pclass',data=df)
# plt.xlabel='survived'
# plt.ylabel='count'
# plt.show()




```
