```python
import pandas as pd

# Load the CSV file
file_path = r"C:\Users\Amukelani\Downloads\male\API_SP.POP.TOTL.MA.ZS_DS2_en_csv_v2_742745.csv"
df = pd.read_csv(file_path, skiprows=4)  # Skipping the first 4 rows as they contain metadata

# Inspect the data
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
# Drop the unnecessary column
df_cleaned = df.drop(columns=['Unnamed: 68'])

# Drop rows with missing values for the year 2019
df_cleaned = df_cleaned.dropna(subset=['2019'])

# Extract relevant columns for visualization
df_visualize = df_cleaned[['Country Name', '2019']]
```

```python
import matplotlib.pyplot as plt

# Sort the data for better visualization
df_visualize = df_visualize.sort_values(by='2019', ascending=False).head(30)  # Top 20 countries

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.barh(df_visualize['Country Name'], df_visualize['2019'], color='skyblue')
plt.xlabel('Percentage of Male Population')
plt.ylabel('Country')
plt.title('Top 20 Countries by Percentage of Male Population in 2019')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest percentage at the top
plt.show()
```
