```python
# read the csv file
import pandas as pd
file_path=r"C:\Users\Amukelani\Downloads\female\API_SP.POP.TOTL.FE.ZS_DS2_en_csv_v2_810737.csv"
df=pd.read_csv(file_path, skiprows=4)
# print(df)
print(df.head())
# print(df.to_string())
print(df.info())
print(df.isnull().sum())
```

```python
#clean data
df_cleaned=df.drop(columns=['Unnamed: 68'])
df_cleaned=df_cleaned.dropna(subset=['2019'])
df_visualize=df_cleaned[['Country Name', '2019']]
```

```python
#data visualization
import matplotlib.pyplot as plt
df_visualize=df_visualize.sort_values(by='2019').head(20)
plt.figure(figsize=(12,10))
plt.barh(df_visualize['Country Name'], df_visualize['2019'])
plt.xlabel('Percentage of Female Population')
plt.ylabel('Country')
plt.title('Distribution of Females in different countries')
plt.show()

         
           
```
